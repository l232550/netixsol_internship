# Business Qsts

## Q1. Which 5 regions had the highest population in 2017?

```python
highest_population_2017 = df.sort_values("population_2017",ascending=False).head(5)
highest_population_2017[["region","population_2017"]]
```

## Q2. Which 5 regions had the lowest population in 2017?

```python
lowest_population_2017 = df.sort_values("population_2017").head(5)
lowest_population_2017[["region","population_2017"]]
```

## Q3. How many regions had a growth rate above 2.5% in 2017?

```python
greater_than_2pt5 = df["growth_2017"] > 2.5
print(greater_than_2pt5)
```

```python
print(greater_than_2pt5.sum())
```

## Q4. How many regions had a growth rate below 1.8% in 2017?

```python
below_rate_1pt8= df["growth_2017"]<1.8
print(below_rate_1pt8)
```

```python
print(below_rate_1pt8.sum())
```

## Q5. What was the average growth rate in 2017 across all regions?

```python
avg_growth_rate_2017 = df["growth_2017"].mean()
avg_growth_rate_2017 = round(avg_growth_rate_2017,2)
print(avg_growth_rate_2017)
```

## Q6. Which region had the highest growth rate in 2017?

```python
highest_growth_rate_2017 = df.sort_values("growth_2017",ascending=False).head(1)
highest_growth_rate_2017[["region","growth_2017"]]
```

## Q7. Show all regions whose 2017 population exceeded 5,000.

```python
df[df["population_2017"] > 5000][["region","population_2017"]]
```

## Q8. Which regions experienced population growth of more than 3× from 1951 to 2017?

```python
df[df["population_2017"] > df["population_1951"] * 3]["region"]
```
